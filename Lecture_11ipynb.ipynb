```python
import csv  #only works for bba not other dept reason later

def FindNoOFStudentByDept(file, dept):
  count = 0
  for row in file:
    if row ["Program"] == dept:
      count +=1

  return count


try:
  with open("/content/SampleFile.csv", "r") as fi:
    reader = csv.DictReader(fi)

    print(FindNoOFStudentByDept (reader, "BBA"))

except FileNotFoundError:
  print("File does not exist.")

finally:
  fi.close()
```

```python
import csv

def FindStudentID(file): #only unque id jara makeup diche
  StudentId = set()
  for row in file:
    StudentId.add(row['Student Id'])

  return StudentId


try:
  with open("/content/SampleFile.csv") as fi:
    reader = csv.DictReader(fi)

    sid = FindStudentID (reader)
    print(len(sid))
    print(sid)

except FileNotFoundError:
  print("File does not exist.")

finally:
  fi.close()
```
