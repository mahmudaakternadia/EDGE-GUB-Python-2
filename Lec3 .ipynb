```python
password = 'abc123'
MAX_ATTEMPT =3 #cap letter means constant

match = False

i=0

while (i<MAX_ATTEMPT):
  user_pass = input("Enter password: ")
  i+=1

  if(user_pass == password):
    match = True
    break

if (match):
  print('Access Granted!')

else:
  print('Access Denied!!!')
```

```python
a = int(input("Enter digit: "))
b = int(input("Enter any digit until which you want to count: "))

count = 0

while (a>0):
  d = a%10 #count: right to left

  if(d == b):
    break
  count+=1
  a //= 10 # // means if you divide it will give you only int value otherwise python always give float value in dividing result

print(count)

```

```python
a = int(input("Enter a number: "))
b = int(input("Enter the digit you want to count until: "))
string_a = str(a)

count = 0
i=0

while i < len(string_a):  #count left to right
    d = int(string_a[i])
    i+=1

    if d == b:
        break

    count += 1

print("Count of digits until", b, "is found: ", count)

```
